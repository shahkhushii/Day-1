# MATPLOTLIB 

```python
import matplotlib.pyplot as mpl
```

```python
mpl.title("CHART")
mpl.xlabel("X-axis")
mpl.ylabel("Y-axis")

x=[1,2,3,4,5]
y=[6,7,8,9,10]
mpl.plot(x,y)
mpl.show()
```

```python
c='r'
mpl.plot(x,y,c)
```

```python
mpl.plot(x,y,marker='o')
```

```python
a='r'
b='g'
mpl.plot(x,a,marker='o')
mpl.plot(y,b,marker='>')
mpl.plot(x,y)
```

```python
mpl.title("CHART")
mpl.xlabel("X-axis")
mpl.ylabel("Y-axis")
mpl.bar(x,y)
```

```python
c='r','b','g','y','k'
mpl.bar(x,y,color=c)
```
